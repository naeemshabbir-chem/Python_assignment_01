```python
# Taking input from user
print("Enter marks for 5 subjects (out of 100):")

m1 = float(input("Biology: "))
m2 = float(input("Mathematics: "))
m3 = float(input("Physics: "))
m4 = float(input("Chemistry: "))
m5 = float(input("English: "))

# Calculations
total = m1 + m2 + m3 + m4 + m5
percentage = total / 5

# Grade calculation
if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 50:
    grade = "D"
else:
    grade = "F"

# Printing Mark Sheet
print("\n----- MARK SHEET -----")
print(f"Biology: {m1}")
print(f"Mathematics: {m2}")
print(f"Physics: {m3}")
print(f"Chemistry: {m4}")
print(f"English: {m5}")
print("-----------------------")
print(f"Total Marks: {total}")
print(f"Percentage: {percentage:.2f}%")
print(f"Grade: {grade}")
print("-----------------------")
```
